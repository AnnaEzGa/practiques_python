# EXEMPLES DE MATRIUS

```python
A=np.array([[2,4],[1,3]])
print(A)
```

```python
e=A[1,0] # Agafa l'element de la fila 1 i la columna 0 de la matriu
print(e)
```

```python
# Seleccionar una fila o columna sencera a partir d'un vector
v=A[0,:] # ES defineix un vector v que es basa en tota la primera fila (0) i tots els elements, totes les coulmnes d'aquesta fila (:)
print(v)
```

```python
u=A[:,1] # Amb això estem agafant tota la primera columna.
print(u)
```

Anem a diagonalitzar matrius i per això necessitem obtenir els valors pròpis, els vectors pròpis...:

1. Vectors pròpis:

```python
A=np.array([[2,2],[1,3]])
print(A)
```

```python
# Valors de vectors pròpis (si no es sap quina és la comanda necessària, es busca a la pàgina web de Numpy)
Valors,Vectors = np.linalg.eig(A) #Ambs això es calcula els valors i vectors pròpis
print("Valors propis",Valors)
print("Vectors propis",Vectors)
```

```python
u=Vectors[:,0] #Amb això estrec el primer vector pròpi que està a la primera columna
print(u)
```

```python
v=Vectors[:,1]
print(v)
# Amb això estrec el segon vector pròpi
```

```python
z=np.dot(A,u) #Multplicar la matriu A per el vector u. Amb això comprobem que efectivament son vectors pròpis.
print(z)
```

```python
z2=np.dot(A,v) #Multplicar la matriu A per el vector v. Amb això comprobem que efectivament son vectors pròpis.
print(z2)
```
